```python
import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv("/titanic.csv")
titanic.head()
titanic.tail()
```

```python
import seaborn as sns
sns.set(style="ticks",color_codes=True)
titanic.plot(kind="scatter",x="Age",y="Fare")
plt.show()
```

```python
sns.lineplot(x="Pclass",y="Fare",data=titanic)
plt.show()
```

```python
titanic_hist = titanic.Fare.plot.hist(bins = 10, color = 'blue')
plt.xlabel('Fare in pounds')
plt.show(titanic_hist)
```

```python
titanic_pclass_fig, titanic_pclass_ax = plt.subplots()
color = ['black', 'green', 'lightblue']
count = 0
for name, group in titanic.groupby('Pclass'):
  titanic_pclass_ax.plot(group.Age, group.Fare, '.',
label = name, alpha = 0.6,
c = color[count])
count += 1
titanic_pclass_ax.legend(numpoints=1, title = "Passenger class",
fontsize = 10)
plt.xlabel('Age')
plt.ylabel('Fare in pounds')
titanic_pclass_ax.set_xlim(-1, 85)
titanic_pclass_ax.set_ylim(-1, 600)
plt.show(titanic_pclass_fig)
```

```python
pclass_fare_titanic = titanic[['Pclass', 'Fare']].pivot(columns =
'Pclass', values = 'Fare')
box_color = dict(boxes = 'black',
whiskers = 'black',
medians = 'blue',
caps = 'black')
titanic_pclass_boxplot = pclass_fare_titanic.plot.box(color =
box_color)
plt.xlabel('Passenger class')
plt.ylabel('Fare in pounds')
plt.show(titanic_pclass_boxplot)
```
